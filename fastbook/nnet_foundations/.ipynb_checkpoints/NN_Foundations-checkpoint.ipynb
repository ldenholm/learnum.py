```python
# Important properties of neural network architectures:
At a high level these architectures are simply mathematical functions, involving matrix multiplications,
nonlinear activation functions and so on.
- How fast are they?
- How much memory does it use?
- How accurate are they?
```

```python
# Create basic quadratic function builder
def quadratic(a, b, c, x): return a*x**2 + b*x + c

# test
quadratic(1, 4, 5, 2)
```

```python
# Create quadratic function builder with fixed coefficients
# Fix (a, b, c) = (2, 4, 3)
from functools import partial
def create_quad(a, b, c):
    return partial(quadratic, a, b, c)

# test
test_fn = create_quad(1, 4, 5)
test_fn(2)
```

```python
# Now we want to plot this function
from fastbook import *
fn = create_quad(2, 4, 3)
plot_function(fn, "2x^2 + 4x + 3")
fn(2)
```

```python
# Next we want to generate some data that closley resembles this function, a
# prelude to fitting it with this curve.
from numpy.random import normal,seed,uniform
np.random.seed(22)
def noise(x, scale): return normal(scale=scale, size=x.shape)
def add_noise(x, mult, add): return x * (1 + noise(x, mult)) + noise(x, add)

x = torch.linspace(-2, 2, steps=20)[:None]
y = add_noise(fn(x), 0.3, 1.5)
plt.scatter(x,y)
```
